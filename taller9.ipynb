Taller-9

```python
from sklearn.model_selection import train_test_split 

train, test = train_test_split(df, test_size = 0.20)
```

```python
train
```

```python
x = train.iloc[:,[4,5,6]]
y = train.Survived
logit_model = sm.Logit(y,x)
result = logit_model.fit()

resum = result.summary()

print(resum)
```

```python
#edadcoefEdad + sibcoefSibSp + parch*coefParch
import statsmodels.api as sm

results_summary = resum

resultsHTML = results_summary.tables[1].as_html()

table = pd.read_html(resultsHTML, header=0, index_col=0)[0]
```

```python
ageCoef = table['coef'].values[0]
SibSpCoef = table['coef'].values[1]
ParchCoef = table['coef'].values[2]
```

```python
import math
```

```python
truePositive = 0
falseNegative = 0
falsePositive = 0
trueNegative = 0

probList = []

for i in test.index:

  valor = (ageCoef*test['Age'][i]) + (SibSpCoef*test['SibSp'][i]) + (ParchCoef*test['Parch'][i])

  prob = 1 / (1 + math.e ** (-1 * valor))

  probList.append(prob)

  if prob<0.5:
    trunc = 0
  else:
    trunc = 1

  trueVal = test['Survived'][i]

  if(trueVal == 1 and trunc == 1):
    truePositive += 1
  elif(trueVal == 1 and trunc == 0):
    falseNegative += 1
  elif(trueVal == 0 and trunc == 1):
    falsePositive += 1
  elif(trueVal == 0 and trunc == 0):
    trueNegative += 1

print("Verdadero Positivo: "+ str(truePositive))
print("Falso Negativo: " + str(falseNegative))
print("Falso Positivo: " + str(falsePositive))
print("Verdadero Negativo: " + str(trueNegative))

print(probList)
```

```python
p = truePositive / (truePositive + falsePositive)
print("Precision: " + str(p*100))
```

```python
r = truePositive / (falseNegative + truePositive)
print("Exhaustividad: " + str(r*100))
```
